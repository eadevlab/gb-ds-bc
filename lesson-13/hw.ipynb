```python
%matplotlib inline
import pandas as pd
import seaborn as sns
```

```python
ds = sns.load_dataset('titanic')
# ds['alone'] = ds['alone'].map({'True': 1, 'False': 0})
ds.head()
```

```python
ds.shape
```

```python
ds.info
```

```python
ds.describe()
```

```python
sns.heatmap(ds.corr(), annot=True);
```

```python
sns.catplot(data=ds, kind='bar', x='alone', y='alive', hue='sex')
```

```python
sns.barplot(data=ds, x='alive', y='age', hue='sex')
```

```python
sns.boxplot(data=ds, x='pclass', y='age', hue='sex')
```
