2. Найдите матрицу С = A - 3B
 
 $\\
A = \left(
\begin{array}{cccc}

\end{array}
\right), B = \left(
\begin{array}{cccc}

```python
a = np.array([[1,2],[2,-1],[3,0]])
b = np.array([[-1, 1],[1, 2],[0, 0]])
c = a - 3*b
c
```

4. Вычислите определитель матрицы А
 
 $\\
A = \left(
\begin{array}{cccc}
6 & -1 & 4\\
2 & 1 & 10\\
8 & 0 & 2\\
\end{array}
\right)$

```python
a = np.array([[6, -1, 4],[2, 1, 10],[8, 0, 2]])
np.linalg.det(a)
```

5. Вычислите матрицу, обратную матрице А. И покажите, что она действительно является обратной
 
 $\\
A = \left(
\begin{array}{cccc}
3 & 3 & -1\\
4 & 1 & 3\\
1 & -2 & -2\\
\end{array}
\right)$

```python
a = np.matrix('3 3 -1;4 1 3; 1 -2 -2')
np.linalg.inv(a)
```
